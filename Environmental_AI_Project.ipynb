```python
import os
from google.colab import userdata
import google.generativeai as genai

# Configure Gemini API
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))

print("✅ Gemini API ready!")
```

```python
# Create Gemini model (latest version)
model = genai.GenerativeModel('gemini-2.5-flash')

print("✅ Environmental AI Agent created!")
```

```python
# Environmental data analysis
prompt = """
আপনি একজন Environmental AI Assistant।

📍 Location: Dhaka, Bangladesh
🌫️ Air Quality Index (AQI): 150
💨 PM2.5 Level: 55 µg/m³

এই data analyze করে বলুন:
1. Air quality এর অবস্থা কেমন?
2. Health এ কী প্রভাব পড়তে পারে?
3. কী সতর্কতা নেওয়া উচিত?

Bengali এবং English উভয় ভাষায় সংক্ষেপে উত্তর দিন।
"""

response = model.generate_content(prompt)

print("🌍 Environmental Analysis:")
print("="*60)
print(response.text)
```

```python
# Compare multiple cities
cities_data = """
আপনি একজন Environmental AI Assistant।

নিচের শহরগুলোর air quality compare করুন:

1. Dhaka: AQI 150, PM2.5: 55 µg/m³
2. Delhi: AQI 180, PM2.5: 70 µg/m³
3. Tokyo: AQI 50, PM2.5: 15 µg/m³

কোন শহরের বাতাস সবচেয়ে ভালো এবং কোনটা সবচেয়ে খারাপ?
সংক্ষেপে Bengali এবং English এ বলুন।
"""

response = model.generate_content(cities_data)

print("🌍 Cities Comparison:")
print("="*60)
print(response.text)
```

```python
import matplotlib.pyplot as plt

# Data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
aqi = [120, 150, 180, 140, 160, 170, 155]
pm25 = [45, 55, 65, 50, 58, 62, 56]

# Create charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# AQI Chart
ax1.plot(days, aqi, marker='o', color='red', linewidth=2, markersize=8)
ax1.axhline(y=150, color='orange', linestyle='--', label='Unhealthy Threshold')
ax1.set_title('📊 Weekly AQI Trend - Dhaka', fontsize=14, fontweight='bold')
ax1.set_xlabel('Day')
ax1.set_ylabel('AQI Value')
ax1.grid(True, alpha=0.3)
ax1.legend()

# PM2.5 Chart
ax2.bar(days, pm25, color='blue', alpha=0.7)
ax2.axhline(y=55, color='red', linestyle='--', label='WHO Limit')
ax2.set_title('💨 Weekly PM2.5 Levels - Dhaka', fontsize=14, fontweight='bold')
ax2.set_xlabel('Day')
ax2.set_ylabel('PM2.5 (µg/m³)')
ax2.grid(True, alpha=0.3, axis='y')
ax2.legend()

plt.tight_layout()
plt.show()

print("✅ Visualization created!")
```
